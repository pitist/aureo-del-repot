## Integrar Base de Datos

### Subtask:
Seleccionar y configurar una base de datos (por ejemplo, SQLite para prototipos, PostgreSQL/MySQL para producción) que se adapte a las necesidades del proyecto. Diseñar el esquema inicial de la base de datos, incluyendo tablas para usuarios, permisos y datos de la aplicación.


### Paso 1: Instalar SQLite (módulo `sqlite3` de Python)

El módulo `sqlite3` de Python es una parte estándar de la biblioteca de Python y generalmente no requiere una instalación separada con `pip`. Ya debería estar disponible en su entorno de Colab.

Podemos verificar su disponibilidad y pasar directamente a la conexión a la base de datos.

**Reasoning**:
Now that it's clarified that `sqlite3` is a built-in module, the next step is to implement the connection to the SQLite database as per the instructions. This will involve importing `sqlite3` and defining a function to establish a connection to `pistix.db`.



```python
import sqlite3
import os

def connect_db(db_name='pistix.db'):
    """Establece una conexión a la base de datos SQLite."""
    conn = None
    try:
        conn = sqlite3.connect(db_name)
        print(f"Conexión exitosa a la base de datos: {db_name}")
        return conn
    except sqlite3.Error as e:
        print(f"Error al conectar a la base de datos {db_name}: {e}")
    return conn

# Establecer la conexión a la base de datos
db_connection = connect_db()

# Verificar si la conexión fue exitosa y cerrar si no es el siguiente paso
if db_connection:
    print(f"La base de datos '{db_connection.execute('PRAGMA database_list;').fetchone()[2]}' ha sido creada o ya existe.")
    # In a real scenario, you'd keep the connection open for schema creation,
    # but for this verification step, we can close it for now if not immediately used.
    # It will be re-opened for table creation.
    db_connection.close()
    print("Conexión verificada y cerrada temporalmente.")
else:
    print("No se pudo establecer la conexión a la base de datos.")
```

**Reasoning**:
With the database connection verified, the next step is to create the specified `users` and `permissions` tables. I will reuse the `connect_db` function, define SQL `CREATE TABLE` statements with the given schema, execute them, commit the changes, and then close the connection.



```python
print("Creando tablas iniciales: users y permissions...")

def create_tables(db_name='pistix.db'):
    """Crea las tablas users y permissions en la base de datos."""
    conn = connect_db(db_name) # Re-establecer la conexión
    if conn:
        try:
            cursor = conn.cursor()

            # Tabla users
            cursor.execute("""
                CREATE TABLE IF NOT EXISTS users (
                    id INTEGER PRIMARY KEY AUTOINCREMENT,
                    username TEXT UNIQUE NOT NULL,
                    password_hash TEXT NOT NULL,
                    email TEXT UNIQUE
                );
            """)
            print("Tabla 'users' creada o ya existente.")

            # Tabla permissions
            cursor.execute("""
                CREATE TABLE IF NOT EXISTS permissions (
                    id INTEGER PRIMARY KEY AUTOINCREMENT,
                    name TEXT UNIQUE NOT NULL,
                    description TEXT
                );
            """)
            print("Tabla 'permissions' creada o ya existente.")

            conn.commit()
            print("Tablas creadas y cambios guardados.")
        except sqlite3.Error as e:
            print(f"Error al crear tablas: {e}")
        finally:
            conn.close()
            print("Conexión a la base de datos cerrada después de crear tablas.")
    else:
        print("No se pudo establecer la conexión para crear tablas.")

# Ejecutar la creación de tablas
create_tables()
```

```python
import os

print("1.2. Updating rust_lib/src/lib.rs to import argon2 and add password hashing/verification functions...")

lib_rs_path = os.path.join(os.getcwd(), 'rust_lib', 'src', 'lib.rs')

with open(lib_rs_path, 'w') as f:
    f.write("""
use argon2::{Algorithm, Argon2, Params, Version};
use rand::rngs::OsRng;
use rand::RngCore;
use uuid::Uuid;

pub fn greet_from_rust(name: &str) -> String {
    format!("Hello, {} from Rust!", name)
}

pub fn generate_uuid_from_rust() -> String {
    Uuid::new_v4().to_string()
}

pub fn hash_password_from_rust(password: &str) -> Result<String, String> {
    let mut salt = [0u8; 16];
    OsRng.fill_bytes(&mut salt);

    let params = Params::new(10240, 2, 1, Some(32)) // memory, iterations, parallelism, output_len
        .map_err(|e| format!("Failed to create argon2 params: {}", e.to_string()))?;
    let argon2 = Argon2::new(Algorithm::Argon2id, Version::V0_19, params);

    argon2.with_salt(&salt)
        .hash_password(password.as_bytes())
        .map_err(|e| format!("Failed to hash password: {}", e.to_string()))
        .map(|hash| hash.to_string())
}

pub fn verify_password_from_rust(password: &str, hash: &str) -> Result<bool, String> {
    let argon2 = Argon2::default();
    argon2.verify_password(password.as_bytes(), hash)
        .map_err(|e| format!("Failed to verify password: {}", e.to_string()))
}

#[cfg(test)]
mod tests {
    use super::*;

    #[test]
    fn it_works() {
        assert_eq!(greet_from_rust("World"), "Hello, World from Rust!");
    }

    #[test]
    fn test_generate_uuid() {
        let uuid_str = generate_uuid_from_rust();
        assert!(!uuid_str.is_empty());
        // Basic check for UUID format (length and hyphens)
        assert_eq!(uuid_str.len(), 36);
        assert_eq!(uuid_str.chars().filter(|&c| c == '-').count(), 4);
    }

    #[test]
    fn test_hash_and_verify_password() {
        let password = "my_secret_password";
        let hashed_password = hash_password_from_rust(password).expect("Failed to hash password");
        println!("Hashed password: {}", hashed_password);
        assert!(!hashed_password.is_empty());
        assert_ne!(password, &hashed_password);

        let is_valid = verify_password_from_rust(password, &hashed_password).expect("Failed to verify password");
        assert!(is_valid, "Password verification failed for correct password");

        let is_invalid = verify_password_from_rust("wrong_password", &hashed_password).expect("Failed to verify wrong password");
        assert!(!is_invalid, "Password verification succeeded for incorrect password");

        // Test verification with an invalid hash format
        let invalid_hash_result = verify_password_from_rust(password, "invalid_hash");
        assert!(invalid_hash_result.is_err());
        assert!(invalid_hash_result.unwrap_err().contains("Invalid hash format"));
    }
}
""")
print("Argon2 import and password hashing/verification functions added to rust_lib/src/lib.rs.")
```
