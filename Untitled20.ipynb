````python
%%bash

set -e # Exit immediately if a command exits with a non-zero status.

# Variables
APP_NAME=AureaApp
PKG=com.aureo.app
SDK_API=34
BUILD_TOOLS=34.0.0
GRADLE_VERSION=8.4
ANDROID_SDK_ROOT=$HOME/android-sdk
GRADLE_HOME=$HOME/gradle # Define GRADLE_HOME

# Export environment variables for the entire script
export ANDROID_SDK_ROOT
export ANDROID_HOME=$ANDROID_SDK_ROOT # Essential for some Android tools and Gradle
# Update PATH to include Gradle bin directory
export PATH=$ANDROID_SDK_ROOT/cmdline-tools/latest/bin:$ANDROID_SDK_ROOT/platform-tools:$GRADLE_HOME/gradle-${GRADLE_VERSION}/bin:$PATH

# Convert package name to path format
PKG_PATH=$(echo $PKG | tr '.' '/')

echo "--- Starting Android SDK & Gradle Setup and APK Build ---"

#################################
# 1. Instalar dependencias del sistema
#################################
echo "1. Installing system dependencies..."
sudo apt-get update -qq
sudo apt-get install -y unzip wget openjdk-17-jdk git
echo "System dependencies installed."

#################################
# 2. Configurar Android SDK
#################################
echo "2. Configuring Android SDK..."
mkdir -p "$ANDROID_SDK_ROOT/cmdline-tools"

CMDLINE_TOOLS_ZIP="commandlinetools-linux-11076708_latest.zip"
wget -q "https://dl.google.com/android/repository/${CMDLINE_TOOLS_ZIP}" -O /tmp/${CMDLINE_TOOLS_ZIP}
unzip -qo /tmp/${CMDLINE_TOOLS_ZIP} -d "$ANDROID_SDK_ROOT/cmdline-tools"

# Eliminar el directorio 'latest' existente y crear el nuevo
rm -rf "$ANDROID_SDK_ROOT/cmdline-tools/latest"
mv "$ANDROID_SDK_ROOT/cmdline-tools/cmdline-tools" "$ANDROID_SDK_ROOT/cmdline-tools/latest"
echo "Android SDK Command Line Tools installed."

# Aceptar todas las licencias automáticamente
echo "Accepting Android SDK licenses..."
yes | sdkmanager --licenses
echo "Android SDK licenses accepted."

# Instalar herramientas necesarias
echo "Installing Android SDK components (platform-tools, platforms;android-${SDK_API}, build-tools;${BUILD_TOOLS})..."
sdkmanager "platform-tools" "platforms;android-${SDK_API}" "build-tools;${BUILD_TOOLS}"
echo "Android SDK components installed."

#################################
# 3. Configurar Gradle
#################################
echo "3. Configuring Gradle..."
mkdir -p "$GRADLE_HOME"

GRADLE_ZIP="gradle-${GRADLE_VERSION}-bin.zip"
wget -q "https://services.gradle.org/distributions/${GRADLE_ZIP}" -O /tmp/${GRADLE_ZIP}
unzip -qo /tmp/${GRADLE_ZIP} -d "$GRADLE_HOME"
echo "Gradle ${GRADLE_VERSION} installed to $GRADLE_HOME."

#################################
# 4. Crear Proyecto Android Mínimo
#################################
echo "4. Creating minimum Android project structure for $APP_NAME..."
rm -rf "$APP_NAME" # Clean up any previous project
mkdir -p "$APP_NAME/app/src/main/java/${PKG_PATH}"
mkdir -p "$APP_NAME/app/src/main/res/layout"
mkdir -p "$APP_NAME/gradle/wrapper"

cd "$APP_NAME"
echo "Changed directory to $(pwd)"

# settings.gradle
cat > settings.gradle <<EOF
rootProject.name = "$APP_NAME"
include ':app'
EOF
echo "settings.gradle created."

# build.gradle (proyecto raíz)
cat > build.gradle <<EOF
buildscript {
    repositories {
        google()
        mavenCentral()
    }
    dependencies {
        classpath 'com.android.tools.build:gradle:8.1.0' # Using a compatible AGP version
    }
}
allprojects {
    repositories {
        google()
        mavenCentral()
    }
}
EOF
echo "Root build.gradle created."

# build.gradle (módulo de la aplicación)
mkdir -p app # Ensure app directory exists before writing app/build.gradle
cat > app/build.gradle <<EOF
plugins {
    id 'com.android.application'
}

android {
    namespace "$PKG"
    compileSdk $SDK_API

    defaultConfig {
        applicationId "$PKG"
        minSdk 26
        targetSdk $SDK_API
        versionCode 1
        versionName "1.0"
    }
    buildTypes {
        release {
            minifyEnabled false
        }
    }
}
dependencies {
    implementation 'androidx.core:core-ktx:1.12.0' # Example dependency
}
EOF
echo "App build.gradle created."

#################################
# 5. Generar local.properties
#################################
echo "5. Generating local.properties..."
echo "sdk.dir=$ANDROID_SDK_ROOT" > local.properties
echo "local.properties created with sdk.dir=$ANDROID_SDK_ROOT."

#################################
# 6. Configurar gradle.properties
#################################
echo "6. Configuring gradle.properties for AndroidX..."
cat > gradle.properties <<EOF
android.useAndroidX=true
android.enableJetifier=true
EOF
echo "gradle.properties created."

# AndroidManifest.xml
mkdir -p app/src/main # Ensure app/src/main exists
cat > app/src/main/AndroidManifest.xml <<EOF
<manifest xmlns:android="http://schemas.android.com/apk/res/android"
    package="$PKG">

    <uses-permission android:name="android.permission.INTERNET"/>

    <application
        android:label="Áurea"
        android:icon="@android:drawable/ic_menu_compass">

        <activity android:name=".MainActivity">
            <intent-filter>
                <action android:name="android.intent.action.MAIN"/>
                <category android:name="android.intent.category.LAUNCHER"/>
            </intent-filter>
        </activity>

    </application>
</manifest>
EOF
echo "AndroidManifest.xml created."

# MainActivity.java
mkdir -p app/src/main/java/${PKG_PATH} # Ensure path exists
cat > app/src/main/java/${PKG_PATH}/MainActivity.java <<EOF
package $PKG;

import android.app.Activity;
import android.os.Bundle;
import android.widget.TextView;

public class MainActivity extends Activity {
    @Override
    protected void onCreate(Bundle savedInstanceState) {
        super.onCreate(savedInstanceState);
        TextView tv = new TextView(this);
        tv.setText("Sistema Áureo Operativo");
        setContentView(tv);
    }
}
EOF
echo "MainActivity.java created."

#################################
# 7. Generar Gradle Wrapper
#################################
echo "7. Generating Gradle wrapper..."
# Ensure the full path to gradle is used to avoid PATH issues for this specific command
"$GRADLE_HOME/gradle-${GRADLE_VERSION}/bin/gradle" wrapper --gradle-version "${GRADLE_VERSION}"

if [ ! -f gradlew ]; then
    echo "Error: gradlew script was not generated by Gradle wrapper command."
    exit 1
fi
chmod +x gradlew
echo "Gradle wrapper generated and made executable."
ls -l gradlew # Verify gradlew exists and has execute permissions

#################################
# 8. Compilar APK
#################################
echo "8. Compiling APK using ./gradlew assembleDebug..."
./gradlew assembleDebug

if [ $? -ne 0 ]; then
    echo "Error: APK compilation failed."
    exit 1
fi
echo "APK compilation successful."

# Mover APK a carpeta pública para descarga
echo "Moving APK to public download folder..."
mkdir -p /content/download
APK_PATH="app/build/outputs/apk/debug/app-debug.apk"

if [ ! -f "$APK_PATH" ]; then
    echo "Error: Generated APK not found at expected path: $APK_PATH"
    exit 1
fi
cp "$APK_PATH" /content/download/
echo "APK '$APK_PATH' copied to /content/download/app-debug.apk."

echo "--- Android SDK & Gradle Setup and APK Build Completed ---"
echo "APK generado y disponible en: /content/download/app-debug.apk"
````

### Cómo instalar y ejecutar el APK en un emulador de Android

Para instalar y ejecutar el APK en un emulador, sigue estos pasos:

1.  **Asegúrate de tener Android Studio instalado**: Si no lo tienes, puedes descargarlo desde el sitio web oficial de Android Developers. Android Studio incluye el Android SDK y el Android Virtual Device (AVD) Manager, que son necesarios para crear y gestionar emuladores.

2.  **Crea un emulador (Android Virtual Device - AVD)**:

    *   Abre Android Studio.
    *   Ve a `Tools > AVD Manager`.
    *   Haz clic en `Create Virtual Device`.
    *   Selecciona una definición de hardware (por ejemplo, `Pixel 6`).
    *   Selecciona una imagen de sistema (Android version), por ejemplo, `API 34`. Descárgala si es necesario.
    *   Haz clic en `Next` y luego en `Finish` para crear el AVD.

3.  **Inicia el emulador**:

    *   En el `AVD Manager`, busca el emulador que acabas de crear y haz clic en el botón de "Play" (triángulo verde) para iniciarlo.
    *   Espera a que el emulador arranque completamente.

4.  **Descarga el APK a tu máquina local**:

    *   En Colab, descarga el archivo `app-debug.apk` desde la ruta `/content/download/` a tu computadora local. Puedes hacerlo haciendo clic en el icono de la carpeta a la izquierda, navegando a esa ubicación y haciendo clic derecho sobre el archivo para descargarlo.

5.  **Instala el APK usando ADB (Android Debug Bridge)**:

    *   Abre una terminal o línea de comandos en tu computadora.
    *   Navega a la carpeta donde descargaste el `app-debug.apk`.
    *   Asegúrate de que `adb` esté en tu PATH. `adb` se encuentra generalmente en `Android/sdk/platform-tools/`. Si no está en tu PATH, puedes navegar a esa carpeta o proporcionar la ruta completa a `adb`.
    *   Ejecuta el siguiente comando para instalar el APK en el emulador que está en ejecución:

        ```bash
        adb install app-debug.apk
        ```

        Si tienes varios dispositivos/emuladores conectados, es posible que necesites especificar el emulador:

        ```bash
        adb devices
        # Esto listará los dispositivos conectados, por ejemplo: emulator-5554
        adb -s emulator-5554 install app-debug.apk
        ```

    *   Deberías ver un mensaje como `Success` en la terminal.

6.  **Ejecuta la aplicación en el emulador**:

    *   Una vez instalado, la aplicación aparecerá en el cajón de aplicaciones del emulador.
    *   Puedes buscar la aplicación "Áurea" y hacer clic en su icono para iniciarla.

````python
%%bash

set -e # Exit immediately if a command exits with a non-zero status.

# Variables
APP=AureaApp
PKG=com.aureo.app
SDK=34
GRADLE_VERSION=8.4
ANDROID_SDK_ROOT=$HOME/android-sdk
GRADLE_HOME=$HOME/gradle

# Export environment variables (already set by c71e0831, but good for self-contained execution)
export ANDROID_SDK_ROOT
export ANDROID_HOME=$ANDROID_SDK_ROOT
export PATH=$ANDROID_SDK_ROOT/cmdline-tools/latest/bin:$ANDROID_SDK_ROOT/platform-tools:$GRADLE_HOME/gradle-${GRADLE_VERSION}/bin:$PATH

# Convert package name to path format
PKG_PATH=$(echo $PKG | tr '.' '/')

echo "--- PASO 2: Creación del Proyecto y Compilación del APK ---"

# =======================================
# 1. Crear Proyecto Android Mínimo
# =======================================
echo "1. Creating minimum Android project structure for $APP..."
rm -rf "$APP" # Clean up any previous project
mkdir -p "$APP/app/src/main/java/${PKG_PATH}"
mkdir -p "$APP/app/src/main/res/layout"
mkdir -p "$APP/gradle/wrapper"

cd "$APP"
echo "Changed directory to $(pwd)"

# settings.gradle
cat > settings.gradle <<EOG
rootProject.name="AureaApp"
include(":app")
EOG
echo "settings.gradle created."

# build.gradle (proyecto raíz)
cat > build.gradle <<EOG
buildscript {
    repositories {
        google()
        mavenCentral()
    }
    dependencies {
        classpath 'com.android.tools.build:gradle:8.1.0' # Using a compatible AGP version
    }
}
allprojects {
    repositories {
        google()
        mavenCentral()
    }
}
EOG
echo "Root build.gradle created."

# build.gradle (módulo de la aplicación)
mkdir -p app # Ensure app directory exists before writing app/build.gradle
cat > app/build.gradle <<EOG
plugins {
    id 'com.android.application'
}

android {
    namespace '$PKG'
    compileSdk $SDK

    defaultConfig {
        applicationId "$PKG"
        minSdk 26
        targetSdk $SDK
        versionCode 1
        versionName "1.0"
    }
}
dependencies {
    implementation 'androidx.core:core-ktx:1.12.0' # Example dependency
}
EOG
echo "App build.gradle created."

# =======================================
# 2. Generar local.properties
# =======================================
echo "2. Generating local.properties..."
echo "sdk.dir=$ANDROID_SDK_ROOT" > local.properties
echo "local.properties created with sdk.dir=$ANDROID_SDK_ROOT."

# =======================================
# 3. Configurar gradle.properties
# =======================================
echo "3. Configuring gradle.properties for AndroidX..."
cat > gradle.properties <<EOG
android.useAndroidX=true
android.enableJetifier=true
EOG
echo "gradle.properties created."

# AndroidManifest.xml
mkdir -p app/src/main # Ensure app/src/main exists
cat > app/src/main/AndroidManifest.xml <<EOG
<manifest xmlns:android="http://schemas.android.com/apk/res/android"
    package="$PKG">

    <uses-permission android:name="android.permission.INTERNET"/>

    <application
        android:label="Áurea"
        android:icon="@android:drawable/ic_menu_compass">

        <activity android:name=".MainActivity">
            <intent-filter>
                <action android:name="android.intent.action.MAIN"/>
                <category android:name="android.intent.category.LAUNCHER"/>
            </intent-filter>
        </activity>

    </application>
</manifest>
EOG
echo "AndroidManifest.xml created."

# MainActivity.java
mkdir -p app/src/main/java/${PKG_PATH} # Ensure path exists
cat > app/src/main/java/${PKG_PATH}/MainActivity.java <<EOG
package $PKG;

import android.app.Activity;
import android.os.Bundle;
import android.widget.TextView;

public class MainActivity extends Activity {
    @Override
    protected void onCreate(Bundle savedInstanceState) {
        super.onCreate(savedInstanceState);
        TextView tv = new TextView(this);
        tv.setText("¡Hola Áurea, esto es Colab!");
        setContentView(tv);
    }
}
EOG
echo "MainActivity.java created."

# =======================================
# 4. Generar Gradle Wrapper
# =======================================
echo "4. Generating Gradle wrapper..."
"$GRADLE_HOME/gradle-${GRADLE_VERSION}/bin/gradle" wrapper --gradle-version "${GRADLE_VERSION}"

if [ ! -f gradlew ]; then
    echo "Error: gradlew script was not generated by Gradle wrapper command."
    exit 1
fi
chmod +x gradlew
echo "Gradle wrapper generated and made executable."
ls -l gradlew # Verify gradlew exists and has execute permissions

# =======================================
# 5. Compilar APK
# =======================================
echo "5. Compiling APK using ./gradlew assembleDebug..."
./gradlew assembleDebug

if [ $? -ne 0 ]; then
    echo "Error: APK compilation failed."
    exit 1
fi
echo "APK compilation successful."

# Mover APK a carpeta pública para descarga
echo "Moving APK to public download folder..."
mkdir -p /content/download
APK_PATH="app/build/outputs/apk/debug/app-debug.apk"

if [ ! -f "$APK_PATH" ]; then
    echo "Error: Generated APK not found at expected path: $APK_PATH"
    exit 1
fi
cp "$APK_PATH" /content/download/
echo "APK '$APK_PATH' copied to /content/download/app-debug.apk."

echo "--- Android App Build Completed ---"
echo "APK generado y disponible en: /content/download/app-debug.apk"
````

The user wants to execute the provided script to create the Android project structure, generate configuration files, create the MainActivity.java file, generate the Gradle wrapper, compile the APK, and move it to the `/content/download/` directory.



````python
%%bash

set -e # Exit immediately if a command exits with a non-zero status.

# Variables
APP=AureaApp
PKG=com.aureo.app
SDK=34
GRADLE_VERSION=8.4
ANDROID_SDK_ROOT=$HOME/android-sdk
GRADLE_HOME=$HOME/gradle

# Export environment variables (already set by c71e0831, but good for self-contained execution)
export ANDROID_SDK_ROOT
export ANDROID_HOME=$ANDROID_SDK_ROOT
export PATH=$ANDROID_SDK_ROOT/cmdline-tools/latest/bin:$ANDROID_SDK_ROOT/platform-tools:$GRADLE_HOME/gradle-${GRADLE_VERSION}/bin:$PATH

# Convert package name to path format
PKG_PATH=$(echo $PKG | tr '.' '/')

echo "--- PASO 2: Creación del Proyecto y Compilación del APK ---"

# =======================================
# 1. Crear Proyecto Android Mínimo
# =======================================
echo "1. Creating minimum Android project structure for $APP..."
rm -rf "$APP" # Clean up any previous project
mkdir -p "$APP/app/src/main/java/${PKG_PATH}"
mkdir -p "$APP/app/src/main/res/layout"
mkdir -p "$APP/gradle/wrapper"

cd "$APP"
echo "Changed directory to $(pwd)"

# settings.gradle
cat > settings.gradle <<EOG
rootProject.name="AureaApp"
include(":app")
EOG
echo "settings.gradle created."

# build.gradle (proyecto raíz)
cat > build.gradle <<EOG
buildscript {
    repositories {
        google()
        mavenCentral()
    }
    dependencies {
        classpath 'com.android.tools.build:gradle:8.1.0' # Using a compatible AGP version
    }
}
allprojects {
    repositories {
        google()
        mavenCentral()
    }
}
EOG
echo "Root build.gradle created."

# build.gradle (módulo de la aplicación)
mkdir -p app # Ensure app directory exists before writing app/build.gradle
cat > app/build.gradle <<EOG
plugins {
    id 'com.android.application'
}

android {
    namespace '$PKG'
    compileSdk $SDK

    defaultConfig {
        applicationId "$PKG"
        minSdk 26
        targetSdk $SDK
        versionCode 1
        versionName "1.0"
    }
}
dependencies {
    implementation 'androidx.core:core-ktx:1.12.0' # Example dependency
}
EOG
echo "App build.gradle created."

# =======================================
# 2. Generar local.properties
# =======================================
echo "2. Generating local.properties..."
echo "sdk.dir=$ANDROID_SDK_ROOT" > local.properties
echo "local.properties created with sdk.dir=$ANDROID_SDK_ROOT."

# =======================================
# 3. Configurar gradle.properties
# =======================================
echo "3. Configuring gradle.properties for AndroidX..."
cat > gradle.properties <<EOG
android.useAndroidX=true
android.enableJetifier=true
EOG
echo "gradle.properties created."

# AndroidManifest.xml
mkdir -p app/src/main # Ensure app/src/main exists
cat > app/src/main/AndroidManifest.xml <<EOG
<manifest xmlns:android="http://schemas.android.com/apk/res/android"
    package="$PKG">

    <uses-permission android:name="android.permission.INTERNET"/>

    <application
        android:label="Áurea"
        android:icon="@android:drawable/ic_menu_compass">

        <activity android:name=".MainActivity">
            <intent-filter>
                <action android:name="android.intent.action.MAIN"/>
                <category android:name="android.intent.category.LAUNCHER"/>
            </intent-filter>
        </activity>

    </application>
</manifest>
EOG
echo "AndroidManifest.xml created."

# MainActivity.java
mkdir -p app/src/main/java/${PKG_PATH} # Ensure path exists
cat > app/src/main/java/${PKG_PATH}/MainActivity.java <<EOG
package $PKG;

import android.app.Activity;
import android.os.Bundle;
import android.widget.TextView;

public class MainActivity extends Activity {
    @Override
    protected void onCreate(Bundle savedInstanceState) {
        super.onCreate(savedInstanceState);
        TextView tv = new TextView(this);
        tv.setText("¡Hola Áurea, esto es Colab!");
        setContentView(tv);
    }
}
EOG
echo "MainActivity.java created."

# =======================================
# 4. Generar Gradle Wrapper
# =======================================
echo "4. Generating Gradle wrapper..."
"$GRADLE_HOME/gradle-${GRADLE_VERSION}/bin/gradle" wrapper --gradle-version "${GRADLE_VERSION}"

if [ ! -f gradlew ]; then
    echo "Error: gradlew script was not generated by Gradle wrapper command."
    exit 1
fi
chmod +x gradlew
echo "Gradle wrapper generated and made executable."
ls -l gradlew # Verify gradlew exists and has execute permissions

# =======================================
# 5. Compilar APK
# =======================================
echo "5. Compiling APK using ./gradlew assembleDebug..."
./gradlew assembleDebug

if [ $? -ne 0 ]; then
    echo "Error: APK compilation failed."
    exit 1
fi
echo "APK compilation successful."

# Mover APK a carpeta pública para descarga
echo "Moving APK to public download folder..."
mkdir -p /content/download
APK_PATH="app/build/outputs/apk/debug/app-debug.apk"

if [ ! -f "$APK_PATH" ]; then
    echo "Error: Generated APK not found at expected path: $APK_PATH"
    exit 1
fi
cp "$APK_PATH" /content/download/
echo "APK '$APK_PATH' copied to /content/download/app-debug.apk."

echo "--- Android App Build Completed ---"
echo "APK generado y disponible en: /content/download/app-debug.apk"
````
